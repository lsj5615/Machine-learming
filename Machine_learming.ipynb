```python
#빙어 몸길이와 몸무게 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

##첫 번쨰 머신러닝 프로그램

>K-최근접 이웃
>>도미와 빙어의 데이터를 합친다(왜 합치는지?)


```python
#두 생선의 데이터 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
#리스트안 리스트로 데이터 정리하기
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
#물고기를 분류해주기 도미(1) 빙어(0)
fish_target = [1]*35 + [0]*14
```

```python
#K-최근접이웃코드 작성(?)
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

```python
#새로운 값에 대한 예측
kn.predict([[30, 600]])
#array([1]) = 도미
```

```python
#주변의 이웃수를 늘리는 코드 ex)49명
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
#이웃을 49명으로 할 시 모두 도미로 예츧하기때문에 정확도가 0.71... 으로 나온다
```

```python
#훈련데이터 지정해주기
train_input = fish_data[:35]
train_target = fish_target[:35]
test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
#훈련 데이터를 가지고 정확도 계산
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn = kn.fit(train_input, train_target)
kn.score(test_input, test_target)
#도미의 데이터로 훈련을 하고 빙어의 데이터로 정확도 측정을해 정확도가 0.0이 나온다
```

#Numpy

```python
#Numpy를 사용하여 열정리
import numpy as np
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print(input_arr)
```

```python
print(input_arr.shape)
#2열 49행을 표시해 준다
```

```python
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
#섞은 숫자를 사용해 생선데이터를 섞기
```

```python
#섞은 데이터를 보여주는 그래프
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

#두 번째 머신러닝

```python
#섞은 데이터를 가지고 정확도 측정
kn = kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
#Numpy를 이용해 생선데이터를 행으로 정리하기
fish_data = np.column_stack((length, weight))
fish_target = np.concatenate((np.ones(35),np.zeros(14)))
```

```python
#생선 테이터를 가지고 Numpy를 사용해 훈련데이터와 테스트 데이터를 만들기
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)
```

```python
#훈련데이터를 가지고 k최근접이웃을 만들기
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
#정확도 100%
```

```python
print(kn.predict([[25, 150]]))
#무게25g, 길이15cm인 생선을 [0.] 빙어로 예측한다(오측)
```

```python
#예측할 데이터 주변의 5개 이웃을 보여주는 그래프 그리기
distances, indexes = kn.kneighbors([[25, 150]])

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
#눈으로 확인해도 주변에는 빙어보단 도미의 데이터가 거 가까운것 같지만 왜 빙어라고 예측을 했던것일까?
```

```python
#x축의 범위를 눈에 보기 쉽게 그래프 그리기
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')

plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
#x축의 범위를 조정하니 길이보단 무게가 물고기의 종을 결정하는데 도움이 되는것 같다
```

#표준점수로 바꾸기

```python
#표준점수로 만들기
mean = np.mean(train_input,axis = 0)
std = np.std(train_input,axis = 0)

print(mean, std)

train_scaled = (train_input - mean) / std
```

```python
new = ([25, 150] - mean) / std
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.fit(train_scaled, train_target)

test_scaled = (test_input - mean) / std
kn.score(test_scaled, test_target)
#1.0
print(kn.predict([new]))
#[1.]
distances, indexes = kn.kneighbors([new])

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
